```python
import matplotlib.pyplot as plt

# Open the text file and read the lines
with open('output3.txt', 'r') as file:
    # Initialize lists to store x and y values
    x_values = []
    y_values = []
    x_values2 = []
    y_values2 = []
    x_values3 = []
    y_values3 = []

    
    # Loop through each line in the file
    for line in file:
        if line.startswith('DL'):
            # Split the line by '/' to extract the components
            parts = line.strip().split('/')
            
            # Extract y value from the second element
            y_value = int(parts[1])
            
            # Calculate x value as the third element multiplied by 1,000,000 plus the fourth element
            x_value = (int(parts[2]) * 1000000 + int(parts[3]))/1000
            # if(int(parts[2]) < 30):
            #     x_value = ((60 + int(parts[2])) * 1000000 + int(parts[3]))/1000
            
            # Append the extracted values to the lists
            if(int(parts[2]) < 55):
                y_values.append(y_value)
                x_values.append(x_value)

        if line.startswith('IS'):

            parts = line.strip().split('/')

            y_value = 5  # y value for IS is default 5
            
            x_value = (int(parts[1]) * 1000000 + int(parts[2]))/1000
            # if(int(parts[1]) < 30):
            #     x_value = ((60 + int(parts[1])) * 1000000 + int(parts[2]))/1000

            if(int(parts[1]) < 55):
                y_values2.append(y_value)
                x_values2.append(x_value)

        if line.startswith('ID'):

            parts = line.strip().split('/')
            
            y_value = 10  # y value for IS is default 5
            
            x_value = (int(parts[1]) * 1000000 + int(parts[2]))/1000
            # if(int(parts[1]) < 30):
            #     x_value = ((60 + int(parts[1])) * 1000000 + int(parts[2]))/1000

            if(int(parts[1]) < 55):
                y_values3.append(y_value)
                x_values3.append(x_value)


plt.figure(figsize=(8, 6))

# Plot DL data
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='DL Lines')

# Plot IS data (x_values2 and y_values2) in a different color
if x_values2 and y_values2:  # Ensure x_values2 and y_values2 are not empty
    plt.plot(x_values2, y_values2, marker='x', linestyle='-', color='r', label='IS Lines')

if x_values3 and y_values3:  # Ensure x_values2 and y_values2 are not empty
    plt.plot(x_values3, y_values3, marker='o', linestyle='-', color='g', label='ID Lines')

# Adding labels and title
plt.xlabel('Time (ms)')
plt.ylabel('Deque Length')
plt.title('Deque Length vs Time')

# Display grid and legend
plt.grid(True)
#plt.legend()

# Show the plot
plt.show()
```

```python
lower_bound1 = x_values[0]  # Set your lower bound
upper_bound1 = x_values2[0]  # Set your upper bound

lower_bound2 = x_values2[20] + 1000  # Set your lower bound
upper_bound2 = x_values2[21]  # Set your upper bound

lower_bound3 = x_values2[40] + 1000
upper_bound3 = x_values2[41]

lower_bound4 = x_values2[len(x_values2) - 1] + 1000
upper_bound4 = x_values[len(x_values) - 1]

in_range_values1 = [x for x in x_values if lower_bound1 < x < upper_bound1]
in_range_values2 = [x for x in x_values if lower_bound2 < x < upper_bound2]
in_range_values3 = [x for x in x_values if lower_bound3 < x < upper_bound3]
in_range_values4 = [x for x in x_values if lower_bound4 < x < upper_bound4]

frequency_count1 = len(in_range_values1)
frequency_count2 = len(in_range_values2)
frequency_count3 = len(in_range_values3)
frequency_count4 = len(in_range_values4)

print("Frequency 1: ", frequency_count1 / (upper_bound1 - lower_bound1))
print("Frequency 2: ", frequency_count2 / (upper_bound2 - lower_bound2))
print("Frequency 3: ", frequency_count3 / (upper_bound3 - lower_bound3))
print("Frequency 4: ", frequency_count4 / (upper_bound4 - lower_bound4))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import colormap

# Set the folder containing the .npy files
data_folder = "20240929data"

# Get a list of all .npy files in the specified folder, sorted alphabetically
npy_files = ([f for f in os.listdir(data_folder) if f.endswith('1.npy')])

# Limit to the first 20 files (if there are fewer, it handles appropriately)
npy_files = npy_files[101:121]

# Set up the number of rows and columns for the subplot grid
num_files = len(npy_files)
rows = 4
cols = 5

# Create a figure with specified size
fig, axes = plt.subplots(rows, cols, figsize=(15, 12))

# Loop through each file and display in the corresponding subplot
for idx, npy_file in enumerate(npy_files):
    # Load the .npy file data
    file_path = os.path.join(data_folder, npy_file)
    data = np.load(file_path)

    # Calculate the position in the subplot grid
    row = idx // cols
    col = idx % cols

    # Display the image with viridis colormap
    ax = axes[row, col]
    ax.imshow(data, cmap='doppler')
    ax.set_title(f"{npy_file}", fontsize=8)
    ax.axis('off')  # Hide axis

# Remove empty subplots if any
for idx in range(num_files, rows * cols):
    fig.delaxes(axes.flatten()[idx])

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```
